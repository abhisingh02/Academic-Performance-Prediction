```python
gender = {
    'male':1,
    'female':0
}
```

```python
race = {
    'group A':0,
    'group B':1,
    'group C':2,
    'group D':3,
    'group E':4
}
```

```python
dataset['gender']=dataset['gender'].map(gender)
dataset['race/ethnicity']=dataset['race/ethnicity'].map(race)
```

```python
dataset
```

```python
level = {
    "bachelor's degree":0,
    'some college':1,
    "master's degree":2,
    "associate's degree":3,
    "high school":4,
    "some high school":5
}
```

```python
dataset['parental level of education']=dataset['parental level of education'].map(level)
```

```python
dataset
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
x = dataset.drop(columns='average_score').values
```

```python
x
```

```python
x[0]
```

```python
y = dataset['average_score'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
predictions
```

```python
from sklearn.metrics import r2_score
```

```python
print(r2_score(predictions,y_test))
```
